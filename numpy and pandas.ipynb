#Numpy Array Operations:  

1. Create a 1D NumPy Array:  

```python
temperatures_w1=np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9] )
print('Temperature for week1 is \n',temperatures_w1)
```

2. Inspection and Properties: 

```python
print('Shape os the array \n',temperatures_w1.shape)
print('Data type of array \n',temperatures_w1.dtype)
print('number of elements in the array \n',temperatures_w1.size)
```

3. Array Operations:  

```python
Fahrenheit=(temperatures_w1*9/5)+32
print('Temperature in Fahrenheit\n',Fahrenheit)

print(f'Maximum temperature for the week is {np.max(Fahrenheit)}F')
print(f'Minimum temperature for the week is {np.min(Fahrenheit)}F')
print(f'Mean temperature for this week is{np.mean(Fahrenheit):.2f}F')
```

4. Array Slicing and Indexing:  

```python
print('Temperature for the week is\n',temperatures_w1)

first_three=temperatures_w1[:3]
last_two=temperatures_w1[-2:]
middle_three=temperatures_w1[2:5]

print('\n Temperature for first three days are\n',first_three)
print('\nTemperature for weekend is\n',last_two)
print('\nTemperature for mid of week\n',middle_three)
```

5. Create a 2D Array:  

```python
temperatures=np.array([[22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9], [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]])

print('temperatures of two weeks:\n',temperatures)
```

6. Inspect and Slice the 2D Array:  


```python
print(f'Shape of the array temperatures is: {temperatures.shape}')
print(f'Data type of the array temperatures is :{temperatures.dtype}')
print(f'Total number of elements in the array : {np.size(temperatures)}')

week1=temperatures[0]
print(f'week1 temperature : {week1}')
week1_end=temperatures[0,-2:]
print(f'First week end temp is: {week1_end}')
week2=temperatures[1]
print(f'Week 2 temperature :  {week2}')
week2_end=temperatures[1,-2:]
print(f'second week end temp : {week2_end}')
```

Pandas Series:  

1. Creating Pandas Series:  

```python
marks=pd.Series([95, 92, 89, 85, 80 ],index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5'  ],name='marks')
marks
```

2. Indexing and Slicing: 

```python
print('mark of first rank student:',marks.iloc[0])
print('\nTop 3 Ranks ')
print(marks.loc[['Rank1','Rank2','Rank3']])
print('\n3rd rank mark: ',marks.iloc[2])

print('\ngreater than 90\n')
print(marks[marks>90])
```

3. Manipulating Series:  

```python
marks['Rank1']=100
```

```python
marks
```

```python
updated_marks=marks.drop('Rank5')
updated_marks
```

```python
CGPA=marks/10
CGPA
```

```python
print('upadted series:\n',updated_marks)
print('\nCGPA series :\n',CGPA)
```
